```python
import apt_pkg
apt_pkg.init_system()

def oudated_packages(tracked, packages):
    for x in packages:
        packages[x] = packages[x].apply(str)
    tracked = (tracked
               .set_index(['package', 'release_snapshot'])
               .merge(packages
                      .set_index(['package', 'release_snapshot'])
                      .rename(columns = {'version':'version_compare'}),
                      left_index=True,
                      right_index=True,
                      how='left')
               .dropna()
               .reset_index()
               .drop_duplicates()
        )
    
    tracked['outdate_3'] = tracked.apply(lambda d:
                                       apt_pkg.version_compare(d['version'],
                                                               d['version_compare']) <0,
                                       axis=1)
    
    tracked = tracked.query('outdate_3 == True').groupby(['package','version']).count()[['outdate_3']].reset_index()
    
    return tracked
```
