```python
import json as js
import os
import codecs
import psycopg2
import pandas as pd

conn_string="host='udd-mirror.debian.net' port='5432' dbname='udd' user='udd-mirror' password='udd-mirror'"

conn = psycopg2.connect(conn_string)

cursor = conn.cursor()

uniq_packages=pd.read_csv('../csv/uniq_packages_docker.csv', sep=';', dtype=object, index_col=None,  error_bad_lines=False)
uniq_packages=uniq_packages.groupby('package').count().reset_index()
f=open('../csv/docker_notProcessed_bugs.csv','w')
g=open('../csv/docker_notProcessed_archived_bugs.csv','w')


all_data=[]
for index, raw in enumerate(uniq_packages.iterrows()):
    print(index)
    package=raw[1]['package']

    cursor.execute(
        "SELECT DISTINCT bugs.id, bugs.status, bugs.severity, "+
        "bugs.arrival, bugs.last_modified, bugs_found_in.version "+
        "FROM bugs, bugs_found_in "+
        "WHERE bugs.id=bugs_found_in.id "+
        "AND bugs.package='"+package+"' ")
    data=cursor.fetchall()
    for x in data:
        id, status, severity, arrival, last_modified, found_in = x
        f.write(package+';'+str(id)+';'+found_in+';'+status+';'+severity+';'+str(arrival)+';'+str(last_modified)+'\n')

    cursor.execute(
    "SELECT archived_bugs.id, archived_bugs.status, archived_bugs.severity, "+
    "archived_bugs.arrival, archived_bugs.last_modified, archived_bugs_found_in.version, archived_bugs_fixed_in.version "+
    "FROM archived_bugs, archived_bugs_found_in, archived_bugs_fixed_in "+
    "WHERE archived_bugs.id=archived_bugs_found_in.id "+
    "WHERE archived_bugs.id=archived_bugs_fixed_in.id "+
    "AND archived_bugs.package='"+package+"'")
    data=cursor.fetchall()
    for x in data:
        id, status, severity, arrival, last_modified, found_in, fixed_in = x
        g.write(package+';'+str(id)+';'+found_in+';'+fixed_in+';'+status+';'+severity+';'+str(arrival)+';'+str(last_modified)+'\n')

g.close()
f.close()
```
