```python
titres=['junit',
        'org.testng',
        'org.hamcrest',
        'org.springframework',
        'org.assertj',
        'org.mockito',
        'org.easymock',
        'org.powermock'
]
pages=['https://mvnrepository.com/artifact/junit/junit',
      'https://mvnrepository.com/artifact/org.testng/testng',
      'https://mvnrepository.com/artifact/org.hamcrest/hamcrest-library',
      'https://mvnrepository.com/artifact/org.springframework/spring-test',
      'https://mvnrepository.com/artifact/org.assertj/assertj-core',
      'https://mvnrepository.com/artifact/org.mockito/mockito-core',
      'https://mvnrepository.com/artifact/org.easymock/easymock',
      'https://mvnrepository.com/artifact/org.powermock/powermock-core']
```

```python
dates=[]
library=[]
versions=[]

for s in range(0,8):
    page = requests.get(pages[s])
    tree = html.fromstring(page.content)
    #title= tree.xpath('//h2[@class="im-title"]/a/text()')
#     date = tree.xpath('//table[@class="grid versions"]/tbody/tr/td/text()')
#     version = tree.xpath('//table[@class="grid versions"]/tbody/tr/td/a[starts-with(@class, "vbtn")]/text()')
    version = tree.xpath('//table[@class="grid versions"]/tbody/tr/td/div/b/text()')
    date = tree.xpath('//table[@class="grid versions"]/tbody/tr[last()]/td/text()')

    date = [x for x in date if x != ' 0\n']
    version = [x for x in version if x != ' 0\n']
    title=titres[s]
    print(title)
    for i in range(0,len(date)):
        tmp=date[i].strip('()')
        tmp=tmp.replace(" ", "")
        tmp=tmp.split(',')
        tmp=tmp[1]+"-"+str(strptime(tmp[0],'%b').tm_mon)+'-01'
        date[i]=tmp
    lib = [title  for x in xrange(len(date))]
    library.extend(lib)
    dates.extend(date)
    versions.extend(version)
library_version = pd.DataFrame(
    {'library': library, 'version': versions, 'date': dates})
```
