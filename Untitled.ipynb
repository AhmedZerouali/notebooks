```python
group=8

f=open('../../valcos/valcos.csv','w')
f.write('id;date;number;prediction;status;group\n')
count=0
j=1
with open('../../valcos/only_number') as lines:
	for index, line in enumerate(lines.readlines()):
		date=line.split('INFO')[0].replace(':\n','')
		number=line.split('--')[0].split(':')[-1].replace(' ','')
		predi=line.split('--')[1].split(':')[1].replace(' ','')
		status=line.split('--')[2].split(':')[1].replace(' ','').rstrip('\n')
		if status=='low':
			continue
		f.write(str(index)+';'+date+';'+str(number)+';'+predi+';'+status+';'+str(count)+'\n')
		if j==group:
			count=count+1
			j=0
		j=j+1
f.close()


numbers=pd.read_csv('../../valcos/valcos.csv', sep=';', dtype=object, index_col='id',  error_bad_lines=False)
numbers.dropna(inplace=True)
numbers['number']=numbers['number'].apply(int)

numbers.groupby('group')['number'].agg(['sum','count']).groupby('sum').count().sort_values(['count'], ascending=False)
```
